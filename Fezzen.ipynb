```python
import glob
import os

# Define the directory containing TIFF files
input_dir = '/content/output/fezzen/content/'

# Define the seasons and the corresponding months
seasons = {
    'winter': ['12', '01', '02'],
    'spring': ['03', '04', '05'],
    'summer': ['06', '07', '08'],
    'autumn': ['09', '10', '11']
}

def get_season_from_filename(filename):
    """Extracts the season based on the filename."""
    basename = os.path.basename(filename)
    print(f"Processing filename: {basename}")  # Debugging line
    try:
        # Extract year-month from filename
        parts = basename.split('-')
        print(f"Parts: {parts}")  # Debugging line
        if len(parts) < 3:
            raise ValueError(f"Unexpected filename format: {basename}")

        # Filename format: WAPOR-3.L3-AETI-D.LCE.2022-04-D1.tif
        month = parts[4]  # '2022-04'
        print(f"Month: {month}")  # Debugging line

        if len(month) != 2:
            raise ValueError(f"Month format issue: {month}")

    except (IndexError, ValueError) as e:
        print(f"Error processing filename '{basename}': {e}")
        return None

    # Determine the season based on the month
    for season, months in seasons.items():
        if month in months:
            return season
    return None

# List all TIFF files in the directory
tiff_files = glob.glob(os.path.join(input_dir, '*.tif'))

# Organize files by season
seasonal_files = {season: [] for season in seasons}
for file in tiff_files:
    season = get_season_from_filename(file)
    if season:
        seasonal_files[season].append(file)

# Print out categorized files
for season, files in seasonal_files.items():
    print(f"Files for {season}:")
    for file in files:
        print(f" - {file}")
```

```python


def sum_tiff_files_by_season(seasonal_files):
    """Sum TIFF files for each season and save the results."""
    for season, file_list in seasonal_files.items():
        if not file_list:
            print(f"No files to process for season: {season}")
            continue

        # Initialize sum array and metadata
        ds_sum = None


        for i, fp in enumerate(file_list):

          ds = rio.open_rasterio(fp)
          ds = ds.where(ds!=ds.attrs['_FillValue'])
          if i == 0:
              ds_sum = ds #Initialize sum if we are looking at the first raster
          else:
              ds_sum += ds #This is the python way of writing ds_sum = ds_sum + ds

        ds_sum.rio.to_raster(season+".tif")



# Sum the TIFF files by season
sum_tiff_files_by_season(seasonal_files)
```

```python
import rasterio
import matplotlib.pyplot as plt

def plot_tiff(file_path, season):
    """Plot a TIFF file with equal aspect ratio."""
    with rasterio.open(file_path) as src:
        # Read the first band
        band1 = src.read(1)

        # Plot
        plt.figure(figsize=(10, 10))
        plt.imshow(band1, cmap='viridis', origin='upper')
        plt.colorbar(label='Value')
        plt.title(f'{season.capitalize()} TIFF')
        plt.xlabel('Column')
        plt.ylabel('Row')

        # Set aspect ratio
        ax = plt.gca()
        ax.set_aspect('equal', adjustable='box')

        plt.show()

# Example usage with seasonal TIFFs
seasonal_tiffs = {
    'winter': '/content/winter.tif',
    'spring': '/content/spring.tif',
    'summer': '/content/summer.tif',
    'autumn': '/content/autumn.tif'
}

for season, file_path in seasonal_tiffs.items():
    print(f"Displaying {season} TIFF...")
    plot_tiff(file_path, season)


```

```python
import geopandas as gpd
import pandas as pd
import rioxarray as rioxr
from rasterstats import zonal_stats
import numpy as np
import matplotlib.pyplot as plt

# Path to seasonal file
summer_AETI = '/content/output/seasonal_tiffs/summer.tif'

# Open the raster file
summer = rioxr.open_rasterio(summer_AETI).squeeze()

# Display the raster file
summer.plot()
plt.show()

# Load your GeoDataFrame (gdf) here
# For example: gdf = gpd.read_file('path_to_your_shapefile.shp')

# Convert CRS to EPSG:32636 if necessary
if gdf.crs != 'epsg:32636':
    gdf = gdf.to_crs(epsg=32636)

# Filter out empty or null geometries
gdf = gdf[~(gdf['geometry'].is_empty | gdf['geometry'].isna())]

# Function to process in batches
def process_in_batches(gdf, raster, batch_size):
    mean_values = []
    num_chunks = len(gdf) // batch_size + (1 if len(gdf) % batch_size > 0 else 0)

    for i in range(num_chunks):
        chunk = gdf.iloc[i * batch_size:(i + 1) * batch_size]
        # Calculate zonal statistics for the chunk
        stats = zonal_stats(
            vectors=chunk,
            raster=raster,
            nodata=np.nan,
            stats="mean",
            all_touched=False
        )
        # Extract mean values and append to the list
        mean_values.extend([stat['mean'] for stat in stats])

    return mean_values

# Process the GeoDataFrame in batches
batch_size = 5  # Adjust this value if necessary
gdf["mean_AETI"] = process_in_batches(gdf, summer_AETI, batch_size)

# Print the updated GeoDataFrame to check the results
print(gdf)
```

```python
# Install required packages
!pip install dask dask-geopandas

# Import necessary libraries
import geopandas as gpd
import pandas as pd
import dask.dataframe as dd
import dask_geopandas as dgpd
import rioxarray as rioxr
from rasterstats import zonal_stats
import numpy as np
import matplotlib.pyplot as plt

# Path to seasonal file
summer_AETI = '/content/output/seasonal_tiffs/summer.tif'

# Open the raster file
summer = rioxr.open_rasterio(summer_AETI).squeeze()

# Display the raster file
summer.plot()
plt.show()

# Load your GeoDataFrame (gdf) here
gdf = gpd.read_file('/content/ly.shp')

# Convert CRS to EPSG:32636 if necessary
if gdf.crs != 'epsg:32636':
    gdf = gdf.to_crs(epsg=32636)

# Filter out empty or null geometries
gdf = gdf[~(gdf['geometry'].is_empty | gdf['geometry'].isna())]

# Convert GeoDataFrame to Dask GeoDataFrame
dgdf = dgpd.from_geopandas(gdf, npartitions=1)  # Adjust the number of partitions as necessary

# Define a function to calculate zonal statistics in chunks
def calculate_zonal_stats(chunk, raster):
    stats = zonal_stats(
        vectors=chunk,
        raster=raster,
        nodata=np.nan,
        stats="mean",
        all_touched=False
    )
    return pd.DataFrame(stats)['mean']

# Apply the function to each partition
mean_values = dgdf.map_partitions(calculate_zonal_stats, raster=summer_AETI, meta=pd.Series(dtype='float64')).compute()

# Assign the mean values to the original GeoDataFrame
gdf['mean_AETI'] = mean_values

# Print the updated GeoDataFrame to check the results
print(gdf)
```
